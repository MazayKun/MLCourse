```python
import numpy as np
from scipy.spatial import distance
import re
import math
import matplotlib.pyplot as plt
```

```python
def tokenize(file_name) :
    with open(file_name, "r") as cat_file: 
        lines = [re.split('[^a-z]', x.lower()) for x in cat_file.readlines()]
    words = []
    i = 0
    for line in lines :
        line = [x for x in line if x != '']
        words.extend(line)
        lines[i] = line
        i+=1
    return words, lines
```

```python
def output(file_name, data) :
    with open(file_name, 'w') as output_file :
        if(type(data) == list):
            for i in data:
                output_file.write(str(i) + ' ')
        else:
            output_file.write(str(data))
```

```python
words, lines = tokenize(r"C:\Users\kir_9\Desktop\sentences.txt")
```

```python
word_freq = []
unique_words = set(words)
i = 0
for line in lines : 
    word_freq.append(dict.fromkeys(unique_words, 0))
    for word in line :
        word_freq[i][word] += 1
    i+=1
```

```python
word_freq = np.array([list(x.values()) for x in word_freq])
cosines = [distance.cosine(x, word_freq[0]) for x in word_freq]
tmp = cosines.copy()
tmp.sort()
output(r"C:\Users\kir_9\Desktop\submission-1.txt", [cosines.index(tmp[1]), cosines.index(tmp[2])])
```

```python
def f(x) :
    return math.sin(x / 5) * math.exp(x / 10) + 5 * math.exp((-1)*x / 2)
```

```python
def poly(x, coefs):
    answ = 0
    tmp = 0
    for i in range(coefs.size):
        tmp = 1
        for j in range(i):
            tmp *= x
        print(str(tmp) + '*' + str(tmp * coefs[i]))
        answ += (tmp * coefs[i])
    return answ
```

```python
F = np.vectorize(f)
x = np.arange(1,15,0.1)
plt.plot(x, F(x))
plt.show()
```

```python
plt.plot(x, [poly(r, np.linalg.solve(np.array([[1,1],[1,15]]), F([1,15]))) for r in x])
```

```python
y = np.linalg.solve(np.array([[1,1,1],[1,8,15],[1,64,15*15]]), F([1,8,15]))
plt.plot(x, [poly(r, y) for r in x])
```
