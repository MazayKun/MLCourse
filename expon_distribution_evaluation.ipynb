```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

# Экспоненциальный закон распределения ($\lambda = 0.5$, $a = 2$, $\sigma = 2$)

```python
expon_rv = sts.expon(scale=1 / 0.5)
```

Сгенерируем выборку объёма 1000 из данного распределения и посторим гистограмму распределения.
На том же графике построим теоретическую плотность распределения.

```python
plt.hist(expon_rv.rvs(1000), density=True)

x = np.linspace(0,10, 100)
pdf = expon_rv.pdf(x)
plt.plot(x, pdf, label='theoretical pdf')
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

Функция, которая рисует гистограмму по заданному объему выборки n.

```python
def hist_sample_mean(rv, n, number_of_samples):
    lst = list()
    for i in range(number_of_samples):    
        sum = 0
        for j in rv.rvs(n):
            sum += j
        lst.append(sum / n)
    plt.hist(lst, density=True)
```

Функция, которая рисует плотность нормального распределения выборочного среднего.

```python
def density_sample_mean(volume, n):
    norm_rv = sts.norm(2, 2 / (n)**0.5)
    x = np.linspace(0,4,volume)
    pdf = norm_rv.pdf(x)
    plt.plot(x, pdf, label='theoretical pdf')
    plt.legend()
    plt.ylabel('$f(x)$')
    plt.xlabel('$x$')
```

```python
# задаем количество выборок
number_of_samples = 1000
```

Для различных объемов выборок n на одном графике строим:
1) гистограмму распределения выборочного среднего;
2) плотность нормального распределения с параметрами: $𝑎=2$, $\sigma = \frac{2}{\sqrt(n)}$



```python
n = 5
hist_sample_mean(expon_rv, n, number_of_samples)
density_sample_mean(100, n)
```

```python
n = 10
hist_sample_mean(expon_rv, n, number_of_samples)
density_sample_mean(100, n)
```

```python
n = 50
hist_sample_mean(expon_rv, n, number_of_samples)
density_sample_mean(100, n)
```

```python
n = 500
hist_sample_mean(expon_rv, n, number_of_samples)
density_sample_mean(100, n)
```

Вывод: С увеличением объема выборки форма гистограммы все лучше повторяет форму плотности нормального распределения.
