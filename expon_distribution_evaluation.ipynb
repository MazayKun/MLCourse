```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

# –≠–∫—Å–ø–æ–Ω–µ–Ω—Ü–∏–∞–ª—å–Ω—ã–π –∑–∞–∫–æ–Ω —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è ($\lambda = 0.5$, $a = 2$, $\sigma = 2$)

```python
expon_rv = sts.expon(scale=1 / 0.5)
```

–°–≥–µ–Ω–µ—Ä–∏—Ä—É–µ–º –≤—ã–±–æ—Ä–∫—É –æ–±—ä—ë–º–∞ 1000 –∏–∑ –¥–∞–Ω–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∏ –ø–æ—Å—Ç–æ—Ä–∏–º –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è.
–ù–∞ —Ç–æ–º –∂–µ –≥—Ä–∞—Ñ–∏–∫–µ –ø–æ—Å—Ç—Ä–æ–∏–º —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫—É—é –ø–ª–æ—Ç–Ω–æ—Å—Ç—å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è.

```python
plt.hist(expon_rv.rvs(1000), density=True)

x = np.linspace(0,10, 100)
pdf = expon_rv.pdf(x)
plt.plot(x, pdf, label='theoretical pdf')
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

–§—É–Ω–∫—Ü–∏—è, –∫–æ—Ç–æ—Ä–∞—è —Ä–∏—Å—É–µ—Ç –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É –ø–æ –∑–∞–¥–∞–Ω–Ω–æ–º—É –æ–±—ä–µ–º—É –≤—ã–±–æ—Ä–∫–∏ n.

```python
def hist_sample_mean(rv, n, number_of_samples):
    lst = list()
    for i in range(number_of_samples):    
        sum = 0
        for j in rv.rvs(n):
            sum += j
        lst.append(sum / n)
    plt.hist(lst, density=True)
```

–§—É–Ω–∫—Ü–∏—è, –∫–æ—Ç–æ—Ä–∞—è —Ä–∏—Å—É–µ—Ç –ø–ª–æ—Ç–Ω–æ—Å—Ç—å –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≤—ã–±–æ—Ä–æ—á–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ.

```python
def density_sample_mean(volume, n):
    norm_rv = sts.norm(2, 2 / (n)**0.5)
    x = np.linspace(0,4,volume)
    pdf = norm_rv.pdf(x)
    plt.plot(x, pdf, label='theoretical pdf')
    plt.legend()
    plt.ylabel('$f(x)$')
    plt.xlabel('$x$')
```

```python
# –∑–∞–¥–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—ã–±–æ—Ä–æ–∫
number_of_samples = 1000
```

–î–ª—è —Ä–∞–∑–ª–∏—á–Ω—ã—Ö –æ–±—ä–µ–º–æ–≤ –≤—ã–±–æ—Ä–æ–∫ n –Ω–∞ –æ–¥–Ω–æ–º –≥—Ä–∞—Ñ–∏–∫–µ —Å—Ç—Ä–æ–∏–º:
1) –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≤—ã–±–æ—Ä–æ—á–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ;
2) –ø–ª–æ—Ç–Ω–æ—Å—Ç—å –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏: $ùëé=2$, $\sigma = \frac{2}{\sqrt(n)}$



```python
n = 5
hist_sample_mean(expon_rv, n, number_of_samples)
density_sample_mean(100, n)
```

```python
n = 10
hist_sample_mean(expon_rv, n, number_of_samples)
density_sample_mean(100, n)
```

```python
n = 50
hist_sample_mean(expon_rv, n, number_of_samples)
density_sample_mean(100, n)
```

```python
n = 500
hist_sample_mean(expon_rv, n, number_of_samples)
density_sample_mean(100, n)
```
